### 2.1 Busca pelos dados

Para este MVP estou utilizando o banco de dados criado pela Microsoft chamado Adventure Works. O banco de dados AdventureWorks oferece suporte a cenários padrão de processamento de transações on-line para um fabricante de bicicletas fictício - Adventure Works Cycles. Os cenários incluem fabricação, vendas, compras, gerenciamento de produtos, gerenciamento de contatos e recursos humanos. Para essa atividade do MVP, focarei o estudo no esquema [SalesLT], que tem uma versão mais leve dos registros de vendas.

O banco foi conectado ao Azure SQL por conexão direta, não sendo necessária outra ação externa.
